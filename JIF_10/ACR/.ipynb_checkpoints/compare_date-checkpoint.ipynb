### Append the new hits_acr column in the previous data

```python
data1 = pd.read_csv("not_enough_credit_JIF_10.csv")
data2 = pd.read_csv("not_enough_credit_JIF_wACR_10.csv")

new = pd.DataFrame()
new['url'] = data2['url']
new['hits_acr'] = data2['hits_acr']

merged = data1.merge(new, how = "inner", on = "url")
print(len(data1))
print(len(data2))
print(len(merged))
```
