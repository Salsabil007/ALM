```python
import pandas as pd
import numpy as np
from scipy.stats import pearsonr
from scipy import stats
```

```python
data1 = pd.read_csv("not_enough_credit_JIF_15.csv")
data2 = pd.read_csv("not_enough_credit_JIF_wACR_n_15.csv")


new = pd.DataFrame()
new['url'] = data2['url']
new['n'] = data2['n']

merged = data1.merge(new, how = "inner", on = "url")
print(len(merged))
```

```python
d = merged.copy(deep = True)
start_after_2010 = []

for ind in d.index:
    if d['startyear'][ind] > 2010:
        start_after_2010.append(d['startyear'][ind])
    else:
        start_after_2010.append(2010)
        
d['start_after_2010'] = start_after_2010

print(len(d[d['start_after_2010']>2010]), " ",len(d[d['start_after_2010']<=2010]))
        
d['avg_per_year'] = d['n']/(2019 - d['start_after_2010'] + 1)
print(d.head(10))
print(d['start_after_2010'].min()," ",d['start_after_2010'].max())
```

```python
d.corr()
```

### Significance of correlation

```python
#pearsonr - Null - the distributions are uncorrelated and noramlly distributed.
#alternative - two-sided - the correlation is non zero

a = stats.pearsonr(d['avg_per_year'], d['hits'])
b = stats.pearsonr(d['avg_per_year'], d['glam'])
print("for hits ", a) #correlation is statistically significant
print("for glam ",b) #correlation is not statistically significant
```
